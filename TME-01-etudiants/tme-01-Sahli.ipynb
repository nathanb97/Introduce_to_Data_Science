Nous allons maintenant écrire les fonctions permettant de calculer des mesures ''basiques'' sur des données unidimensionnelles. Pour cela, nous allons travailler avec la structure <tt>array</tt> de numpy. Pour convertir une colonne de DataFrame en '<tt>array</tt>, voici commment faire:

### Moyenne

<font color="RED" size="+1">**[Q]**</font> **La moyenne**


Ecrire la fonction moyenne(valeurs) qui calcule la moyenne des éléments d'un tableau (en utilisant la boucle ''for'', sans utiliser les fonctions déjà implémentées dans numpy)

```python
def moyenne(tab):
    if len(tab)==0:
        return 0.0
    nb=0
    s=0
    for i in range(0,len(tab)):
        if tab[i]!="NaN":
            s=s+tab[i]
            nb=nb+1
    return s/nb
        
print("la moyenne (qualité moyenne) est :%f dollars" % moyenne(np.unique(prices_pd["MedQ"].values)))     
#np.unique(prices_pd["MedQ"].values)
```

<font color="RED" size="+1">**[Q]**</font> **Calcul d'un effectif**

Prendre un intervalle de discrétisation de taille 20, et calculer l'effectif (sous forme d'un vecteur) du nombre d'états par ''bins''. Dessiner l'histogramme correspondant

```python
#je cré un tableau dont les entiers sont compris entre min et max
#les valeurs de ce tableau seront définit avec un pas de 20
#plus haut j'ai mon tableau de moyenne, où chaque moyenne est associé à un état
#bins=bacs , pour chaque valeur du tableau de moyennes , je regarde où sa valeur se situe dans le tableau
#j'ai besoin de connaitre mon effectif total(50 etats) et quel sont sont mes effectif partiel
#dans le tableau construit avec un pas de 20 : jaurai les valeur suivante 80 - 100-120- etcc ... 659
#j'ai donc plusieurs intervalle de valeurs,pour chaque valeur de moyennes je regarde dans quel interval se situe la moyenne
#que je suis en train de traiter.
# en abcise de l'histo j'ai mes valeurs définit avec un pas de 20, ex: (80,100) --> vecteur de 2 coordonées , représentant
#les valeurs comprise entre 80 et  100
# en orodnné de l'histo jai l'effectif correspondant --> nombre d'état dont la moyenne se situe dans cette tranche

# mon tableau de moyenne est prix_moyens
#je construit ma liste dont les valeurs sont comprise entre min et max du tableau de moyenne , avec un pas de 20

l=[]
i=prix_min
while i < prix_max+20:  # j'aurai pu utiliser np.arange(prix_min,prix_max,20)
    l.append(i)
    i=i+20

#print(l)

#effectif réprésenter par un dico, clé:valeur --> (borne_min,borne_max):effectief partiel de cette intervale de valeur
dico_effectif={}
#initialisation du dictionnaire
for i in range(0,len(l)-1):
    mon_tuple=(l[i],l[i+1])
    dico_effectif[mon_tuple]=0
    
#print(dico_effectif) 

#pour chaque moyenne rencontré je regarde dans quel intervall elle se situe
#pour chaque moyenne , je parcours le dictionnaire pour voir à quelle intervall , elle correspond , dès que je trouve , 
#j'incrémente la clé du dico qui correspond , et je sort de ma boucle

for i in range (0,len(prix_moyens)):
    moy=prix_moyens[i]
    for key,value in dico_effectif.items():
        borne_min=key[0]
        borne_max=key[1]
        if borne_min<=moy<=borne_max:
            dico_effectif[key]=value+1
            break
            
#print(dico_effectif)
#print(len(dico_effectif))
#maintenant je peux contruire une liste effectif où je ne range que la valeur des effectif
#Ainsi qu'une liste  constitué des tuples de bornes
mes_effectifs=[]
mes_intervalle=[]

for key,value in dico_effectif.items():
    mes_effectifs.append(value)
    mes_intervalle.append(key)
    
print(mes_effectifs)
print(mes_intervalle)

# on a désormais les outils nécessaire pour construire l'histograme

plt.rcdefaults()
fig,ax=plt.subplots()
#en ordonnée je met les intervalle
ordonnee=mes_intervalle
#definit le nombre de valeur sur l'axe des abcisses
y_pos=np.arange(len(mes_intervalle))
error=np.random.rand(len(ordonnee)) # cré un tableau de 1 dimension de longueur len(ordonnee) avec des valeurs aléatoire
#cré une barre horizontal représentant un axe
ax.barh(y_pos, mes_effectifs, xerr=error, align='center', color='green', ecolor='black')
#définit les emplacement sur l'axe des y
ax.set_yticks(y_pos)
#definis mes libéles sur l'axe des y
ax.set_yticklabels(ordonnee)
#les étiquettes (libellés) sont lues de haut en bas
ax.invert_yaxis()
ax.set_xlabel("effectif")
ax.set_title("mes effectis pour chacun desintervalles  des prix moyens")






```

<font color="RED" size="+1">**[Q]**</font> **Changement d'échelle**

Prendre maintenant un intervalle de taille 40.

```python
#définir une liste dont les valeurs sont comprises entre moyenne_min , et moyenne_max avec un pas de 40
l=np.arange(prix_min,prix_max+40,40)
#print(len(l)
#j'initialise mon dictionnaire
dico_eff_40={}
for i in range(0,len(l)-1):
    mon_tuple=(l[i],l[i+1])
    dico_eff_40[mon_tuple]=0

#print(dico_eff_40)
#pour chacune des moyennes , je regarde à quel intervalle elle appartient , et je puis j'incrémente de 1 la valeur
#qui correspond à cet intervalle dans le dico

for i in range(0,len(prix_moyens)):
    moy=prix_moyens[i]
    for key,value in dico_eff_40.items():
        borne_min=key[0]
        borne_max=key[1]
        if borne_min<=moy<=borne_max:
            dico_eff_40[key]=value+1
            break
            
#print(dico_eff_40)
# Maintenant je construit ma liste de tuples pour mes intervalles , et ma liste pour les effectifs
effectifs=[]
intervalles=[]
for key,value in dico_eff_40.items():
    effectifs.append(value)
    intervalles.append(key)
    
print(effectifs)
print(intervalles)
#Maintenant jai les outils pour construire mon histogramme

#abcisses : mes effectifs , ordonnes : intervalles de moyennes associé
plt.rcdefaults()
fig,ax = plt.subplots()
ordonne=intervalles
y_pos=np.arange(len(ordonne))
error=np.random.rand(len(ordonne))
#definit un axe , prend en parametre le tableau des valeurs de l'axe des abcisses (les effectifs)
ax.barh(y_pos,effectifs,xerr=error,align='center',ecolor="black")
ax.set_yticks(y_pos) # je pense que c'est pour la graduation sur l'axe des ordonnée , définir le nombre de valeurs attendu sur cet axe
ax.set_yticklabels(intervalles)  # définit les valeur prises sur l'axe des ordonées
ax.invert_yaxis()
ax.set_xlabel("effectif") # définit "le" nom de l'axe des abcisse
ax.set_title("effectif correspondant à chaque intervalle de prix moyen")









```

<font color="RED" size="+1">**[Q]**</font> **Calcul de la densité cumulée**


Calculer la densité cumulée à partir des prix moyens, avec un intervalle de discrétisation donné et la représenter graphiquement. 

La méthode doit rendre deux tableaux: l'abscisse $X$ (les prix moyens possibles entre prix min et prix max), et la densité associée

```python
#densité cumulé : ajouter au fur et à mesure les densité obtenu
#dénsité --> je l'associe ici à la densité d'effectif
# la densité d'effectif se calcule : effectif/longueur de l'intervalle étudié

def min(tab):
    return np.min(tab)

def max(tab):
    return np.max(tab)

def densite_effectif(longueur_intervalle,effectif):
    return effectif/longueur_intervalle;

#je définit une fonction effectif , qui me permet de construire mes deux tableaux : un pour les intervalles (amplitude) et l'autre pour les effectiffs associés

def effectif(tab_prix_moyen,pas_de_discretisation):
    #définir une liste dont les valeurs sont comprises entre moyenne_min , et moyenne_max avec un pas de 40
    l=np.arange(prix_min,prix_max+pas_de_discretisation,pas_de_discretisation)
    #print(len(l)
    #j'initialise mon dictionnaire
    dico_eff={}
    for i in range(0,len(l)-1):
        mon_tuple=(l[i],l[i+1])
        dico_eff[mon_tuple]=0

    #print(dico_eff)
    #pour chacune des moyennes , je regarde à quel intervalle elle appartient , et je puis j'incrémente de 1 la valeur
    #qui correspond à cet intervalle dans le dico

    for i in range(0,len(prix_moyens)):
        moy=prix_moyens[i]
        for key,value in dico_eff.items():
            borne_min=key[0]
            borne_max=key[1]
            if borne_min<=moy<=borne_max:
                dico_eff[key]=value+1
                break
            
    #print(dico_eff)
    # Maintenant je construit ma liste de tuples pour mes intervalles , et ma liste pour les effectifs
    effectifs=[]
    intervalles=[]
    for key,value in dico_eff.items():
        effectifs.append(value)
        intervalles.append(key)

    return (effectifs,intervalles)  # retourne un tuple de valeur

#mon_tuple=effectif(prix_moyens,20)
#print(mon_tuple[0])

#retourne les prix moyens possible et leur densité associé
def densite(tableau_de_prix_moyen,pas_de_discretisation):
    mon_tuple=effectif(tableau_de_prix_moyen,pas_de_discretisation)
    tab_effectif=mon_tuple[0]
    tab_intervalle=mon_tuple[1] # tableau des intervalle de moyennes
    #je vais construire le tableau de densité , où chaque valeur de densité est associé à un intervalle de moyenne et un effectif
    # c'est pratique , ici tous les intervalle ont la même longueur
    densites=[]
    for i in range(0,len(tab_effectif)):
        eff_courant=tab_effectif[i] # je recup l'effectif courant
        densite=densite_effectif(pas_de_discretisation,eff_courant)
        densites.append(densite)
        
    return (densites,tab_intervalle)  # tableau des intervalles avec chacune des densités qui leur sont associés(elles sont stockés dan sun tableau)
    
#print(densite(prix_moyens,20))

#pour calculer la densité cumulé il suffit d'additionner au fur et à mesure les densités sur chaque point des abcisses
def densite_cumulee(tableau_de_prix_moyen,pas_de_discretisation):
    mon_tuple=densite(tableau_de_prix_moyen,pas_de_discretisation)
    densites=mon_tuple[0]
    intervalles=mon_tuple[1]
    for i in range(1,len(densites)):
        densites[i]=densites[i]+densites[i-1]
        
    return (intervalles,densites)

mon_tuple1=densite_cumulee(prix_moyens,20)
tab_intervalles=mon_tuple1[0]
tab_densites_cumule=mon_tuple1[1]
#print(tab_densites_cumule)
#print(tab_intervalles)

def afficher_histo(tab_ordonnee,tab_abcisse,label_de_x,titre):
    #abcisses : mes effectifs , ordonnes : intervalles de moyennes associé
    plt.rcdefaults()
    fig,ax = plt.subplots()
    ordonne=tab_ordonnee
    y_pos=np.arange(len(ordonne))
    error=np.random.rand(len(ordonne))
    #definit un axe , prend en parametre le tableau des valeurs de l'axe des abcisses (les effectifs)
    ax.barh(y_pos,tab_abcisse,xerr=error,align='center',ecolor="black")
    ax.set_yticks(y_pos) # je pense que c'est pour la graduation sur l'axe des ordonnée , définir le nombre de valeurs attendu sur cet axe
    ax.set_yticklabels(ordonne)  # définit les valeur prises sur l'axe des ordonées
    ax.invert_yaxis()
    ax.set_xlabel(label_de_x) # définit "le" nom de l'axe des abcisse
    ax.set_title(titre)


afficher_histo(tab_intervalles,tab_densites_cumule,"densite","densites cumulé des prix moyens")

# Je ne comprend pas pourquoi quand je souhaite mettre mes moyenne en abcisse ca ne fonctionne pas 

```

<font color="RED" size="+1">**[Q]**</font> **Quartiles**

Ecrire la fonction Q(alpha,x,y) qui permet de trouver le quartile(alpha,x,y).
- quartile(0.5,x,densite) correspondra à la médiane. 

Calculer et représenter graphiquement le boxplot.

Remarque: un boxplot en pandas se fait ainsi:
        
    a=pd.DataFrame(prix_moyens)
    a.boxplot()


```python
#le premier quartile est la plus petite valeur v de la série vérifiant qu'au moins 25% des valeur de la série sont <= v
#La mediane est (la valeur centrale de la série) la plus petite valeur v de la série vérifiant qu'au moins 50% des valeurs de la
#série sont <=v
#le troisième quartile est la + petite valeur vde la série  vérifiant qu'au moins 75% des valeurs de la série sont <=v
#le quatrieme quartile est la + grande valeur de la série --> le max 
#alpha déterminera de quel quartile on traite 
#la fonction recoit en parametre le tableau des effectif
# je vois pas x fait référence à quoi
#je peux caluclerle quartile juste avec leffectif

def eff(tab):
    dico_eff={}
    for i in range(0,len(tab)):
        val=tab[i]
        if val not in dico_eff:
            dico_eff[val]=1
        else:
            dico_eff[val]=dico_eff[val]+1
    return dico_eff #contient la valeur et l'effectif qui lui est associé



def Q(alpha,tab):   #tab est une liste de valeur : une série de nombre par exemple
    per=[0.25,0.5,0.75,1]
    if alpha not in per:
        return false #on ne réfère à aucun quartil
    effectifs=eff(tab)
    #print(effectifs)
    valeurs_associes=[] # je cré le tableau des valeurs associé aux effectif
    liste_effectifs=[]
    for key,value in effectifs.items():
        liste_effectifs.append(value)  #liste de valeurs associé aux effectifs
        valeurs_associes.append(key)
    #print(liste_effectifs)
    #print(valeurs_associes)
    effectif_total=len(tab)
    #maintenant je cré le tableau des effectifs cumulés, de cette manière je peux retourver le quartile simplement
    for i in range(1,len(liste_effectifs)):
        liste_effectifs[i]=liste_effectifs[i]+liste_effectifs[i-1]  # liste effectifs est devenue liste_effectif cumulé
    
    
    if alpha==0.25: # premier quartile
        n=effectif_total/4
        n=math.ceil(n) # j'arrondie à lentier superieur
    elif alpha==0.5: # mediane
        n=effectif_total/2
        n=math.ceil(n)  
    elif alpha==0.75: # troisieme quartile:
        n=effectif_total*3/4
        n=math.ceil(n)
    elif alpha==1:
        n=effectif_total # valeur max de la série
    
    for i in range(0,len(liste_effectifs)):
        if liste_effectifs[i]>=n:
            return valeurs_associes[i]  # je retourne la valeur associé à cette effectif 
    


#Exemple

#tab=[35,36,36,36,36,36,38,38,38,38,39,39,39,39,39,39,39,40,40,40]
#val=Q(0.5,tab)
#print(val)
a=pd.DataFrame(prix_moyens)
a.boxplot()


```

<font color="RED" size="+1">**[Q]**</font> **Corrélations**

A l'aide de la fonction moyenne écrite précédemment, écrire une fonction <tt>correlation(x,y)</tt> qui calcule la corrélation lineaire entre deux tableaux Numpy.

```python
#recherche une corélation entre les deux variable High_x et high_y re : "Sachant que High_x est petit , peut s'attendre à ce que "
#High_y soit également petit ? (ou l'inverse)
#il nous faut in indicateur numérique pour nous dire si les valeurs sont corélés ou pas
#Qand on caarctérise une valeur(ici en disant qu'elle est petite) , on le fait par rapport aux valeur, c'est à dire qu'on
# compare le nombre par rapport à une mmoyenne
#prenons x=High_x et y=High_y , pour mesurer si x est + petit que la moyenne on peut calculer (x-moyenne(x))
#cette quantité sera négative si x<moyenn(x) (et positive dans le cas contraire)
#de même on peut calculer y-moyenne(y) pour comparer y à la moyenne des y
#Maintenant on les multiplie : a=(x-moyenne(x))*(y-moyenne(y))
#si x < moyenne(x) et y<moyenne(y) alors on obtitent la multiplcation de deux nombre négatif, qui nous donne un résultat positif
#cest aussi valable si x>moyenn(x) et y>moyenn(y)
# là on a traiter la quantité a pour une seule valeur des 2 tableaux
#Maintenant il faut le faire pour toute 
#Mais si High_y est vraiment petit quand High_x est petit(et inversement) , alors  les a de toute les opérations serant positif
#la moyenne de tous ces a s'obtient par : [somme de tous les (xi-moyenne(x))*(yi-moyenn(y))  ]*  nombre de valeurs traites
# Au contraire si il n'y a pas de corrélation , c-à-d que y n'est pas forcément petit quand x est petit(et inversement), on obtiendra
# tantot des valeur positives pour (y-moyenne(y)) et tantot dse valeur négatives pour(y-moyenn(y))
#Donc les valeurs de a seront tantôt négatives et tantôtpositives
#Dans ce cas la Moyenne de tous les a sera proche de 0 , et il n'y pas de corrélation
#si on obtient une moyenne négative , on dit qu'il y a une corrélation négative , exemple : si x est grand alors y sera petit

def correlation(tab_x,tab_y):
    s=0
    moy_x=moyenne(tab_x)
    moy_y=moyenne(tab_y)
    for i in range(0,len(tab_x)):
        v1=tab_x[i]-moy_x
        v2=tab_y[i]-moy_y
        a=v1*v2
        s=s+a
    if(len(tab_x)!=0):    
        return s/len(tab_x)
    else:
        return 0.0 # je gère les cas d'erreur , 0 peut etre assimilé à une erreur car une corralation proche de 0 indique qu'il n'y a pas de correlation

```
